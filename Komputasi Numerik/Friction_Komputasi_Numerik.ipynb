```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Soal 1
# Fungsi untuk menghitung gaya friksi
def calculate_friction(normal_force, friction_coefficient):
    return normal_force * friction_coefficient
```

```python
# Gaya friksi maksimum
friction_coefficient = 0.2
normal_force_max = 500
friction_max = calculate_friction(normal_force_max, friction_coefficient)
print(f"Gaya friksi maksimum untuk gaya normal {normal_force_max} N adalah {friction_max} N")
```

```python
# Hitung gaya friksi untuk rentang gaya normal
normal_forces = np.arange(0, 1001, 100)  # Gaya normal dari 0 hingga 1000 N dengan interval 100 N
friction_forces = calculate_friction(normal_forces, friction_coefficient)
```

```python
# Plot grafik
plt.figure(figsize=(8, 6))
plt.plot(normal_forces, friction_forces, marker='o', linestyle='-', color='b', label=f'Koefisien Friksi = {friction_coefficient}')
plt.xlabel('Gaya Normal (N)')
plt.ylabel('Gaya Friksi (N)')
plt.title('Hubungan antara Gaya Normal dan Gaya Friksi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Soal 2
g = 9.8  # Percepatan gravitasi (m/s)
m = 10   # Massa benda (kg)
mu_k = 0.2  # Koefisien friksi kinetis
F_dorong = 100  # Gaya dorong (N)
t_total = 10  # Waktu total (detik)
time_steps = 100  # Jumlah langkah waktu untuk simulasi

# Gaya normal
F_normal = m * g

# Gaya friksi
F_friction = mu_k * F_normal

# Resultan gaya
F_resultant = F_dorong - F_friction

# Percepatan
a = F_resultant / m
```

```python
# Simulasi gerakan benda
times = np.linspace(0, t_total, time_steps)
velocities = a * times  # Kecepatan = percepatan * waktu
distances = 0.5 * a * times**2  # Jarak = 0.5 * percepatan * waktu^2

# Visualisasi gerakan benda
plt.figure(figsize=(14, 6))

# Grafik kecepatan vs waktu
plt.subplot(1, 2, 1)
plt.plot(times, velocities, label='Kecepatan (m/s)', color='blue')
plt.xlabel('Waktu (detik)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Kecepatan vs Waktu')
plt.legend()
plt.grid(True)

# Grafik jarak vs waktu
plt.subplot(1, 2, 2)
plt.plot(times, distances, label='Jarak (m)', color='green')
plt.xlabel('Waktu (detik)')
plt.ylabel('Jarak (m)')
plt.title('Jarak vs Waktu')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Hitung jarak yang ditempuh setelah waktu tertentu
time_specified = 5  # Ambil saja, setelah 5 detik
distance_specified = 0.5 * a * time_specified**2
print(f"Jarak yang ditempuh setelah {time_specified} detik adalah {distance_specified} meter")
```
