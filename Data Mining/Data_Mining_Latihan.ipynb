```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

file_path = 'Data Mining Latihan.csv'
df = pd.read_excel(file_path)

display(df)
```

```python
#1. Missing Value
df['Harga (Rp)'] = df['Harga (Rp)'].fillna(df['Harga (Rp)'].mean()).round().astype(int)
df['Rating'] = df['Rating'].fillna(df['Rating'].mean()).round(1)
df['Jumlah Ulasan'] = df['Jumlah Ulasan'].fillna(df['Jumlah Ulasan'].mean()).round().astype(int)
df['Stok'] = df['Stok'].fillna(df['Stok'].mean()).round().astype(int)
df['Jumlah Terjual'] = df['Jumlah Terjual'].fillna(df['Jumlah Terjual'].mean()).round().astype(int)

df['Status Promo'] = df['Status Promo'].fillna(df['Status Promo'].mode()[0])
df['(Populer/Tidak Populer)'] = df['(Populer/Tidak Populer)'].fillna(df['(Populer/Tidak Populer)'].mode()[0])
display(df)
```

```python
#2. Normalisasi
scaler = MinMaxScaler()
df[['Harga (Rp)', 'Rating', 'Jumlah Ulasan', 'Stok', 'Jumlah Terjual']] = scaler.fit_transform(df[['Harga (Rp)', 'Rating', 'Jumlah Ulasan', 'Stok' ,'Jumlah Terjual']])
display(df)
```

```python
#3. Diskritisasi (Stok)
bins = [0, 0.3, 0.6, 0.8, 1]
labels = ['Low', 'Medium', 'High', 'Very High']
df['Stok Category'] = pd.cut(df['Stok'], bins=bins, labels=labels, include_lowest=True)

display(df)
```
